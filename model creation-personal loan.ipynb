```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("personalloan_data.csv")
```

```python
df
```

# Model Saving

```python
X=df.iloc[:,0:9]
y = df["Personal Loan"]

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn import pipeline
from sklearn_pandas import DataFrameMapper

m = DataFrameMapper([(["Income","Family","CCAvg","Education","Mortgage","CD Account","Online",
                      "CreditCard"],StandardScaler())])

main_model = pipeline.Pipeline([('mapper',m),('model',KNeighborsClassifier(n_neighbors=3))])

main_model.fit(X,y)
```

# Save the Model

```python
import pickle
f1 = open("LogisticpersoloanModel.pkl",mode="wb")
pickle.dump(main_model,f1)
f1.close()
```
